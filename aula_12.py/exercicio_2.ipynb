```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv', sep = ';', encoding='latin1')
dados
```

```python
dados_est = dados.sort_values(by = 'estelionato')
dados_est
```

```python
read = np.array(dados['estelionato'])
read
```
