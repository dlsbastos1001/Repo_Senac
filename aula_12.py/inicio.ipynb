```python
import pandas as pd
import numpy as np
```

```python
ty = np.array([12, 15, 17, 20, 22, 25, 28, 30, 35, 40])
q1 = np.percentile(ty, 25)
q2 = np.percentile(ty, 50)
mediana = np.median(ty)
q3 = np.percentile(ty, 75)
print(f'Mediana: {mediana}')
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2): {q2}')
print(f'Terceiro quartil (Q3): {q3}')
```

```python
dados = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv', sep = ';', encoding='latin1')
dados
```

```python
dados_cel = dados.sort_values(by = 'furto_celular')
dados_cel
```

```python
deleg = dados.groupby('cisp')['furto_celular'].sum()
deleg
```

```python
dados_cel = dados.groupby('cisp')['furto_celular'].sum().reset_index()
dados_cel
```

```python
dados_cel_array = np.array(dados_cel['furto_celular'])
dados_cel_array
```
