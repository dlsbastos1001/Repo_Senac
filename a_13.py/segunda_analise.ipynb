```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('BaseDPEvolucaoMensalCisp.csv', sep = ';', encoding = 'latin1')
dados
```

```python
read = np.array(dados['furto_veiculos'])
read
```

```python
read2 = np.array(dados['roubo_veiculo'])
read2
```

```python
deleg = dados.groupby('furto_veiculos')['roubo_veiculo'].sum().reset_index()
deleg
```

```python
d_capital = dados.groupby('cisp')['SOMA'].sum().reset_index()
d_capital
```

```python
dados_cap = dados[dados['munic'] == 'Rio de Janeiro']
dados_cap = dados_cap.sort_values(by = 'furto_veiculos')
dados_cap

```

```python
dados_cap = dados[dados['regiao'] == 'Capital']
dados_cap = dados_cap.sort_values(by = 'roubo_veiculo')
dados_cap
```

```python
ty = np.concatenate((read, read2))
q1 = np.percentile(ty, 25)
q2 = np.percentile(ty, 50)
mediana = np.median(ty)
q3 = np.percentile(ty, 75)
print(f'Mediana: {mediana}')
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2): {q2}')
print(f'Terceiro quartil (Q3): {q3}')
print(f'media: {round(np.mean(ty), 2)}')
```
